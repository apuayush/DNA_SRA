```python
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = io.imread('../dataset/angle.png')
print(img.shape)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap=plt.cm.gray)
```

```python
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1.2, 100)
print(circles.shape)
x, y, r = circles.flatten()
cir = cv2.circle(img.copy(), (x, y), r, (0, 255, 0), 4)
plt.imshow(cir)
```
